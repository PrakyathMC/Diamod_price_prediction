## EDA

#### Univariate Analysis

```python
fig, axes = plt.subplots(2,2, figsize=(16,10))
fig.suptitle("Distribution of Continous variables")

sns.histplot(df['price'], kde=True, ax=axes[0,0], bins=30 )
axes[0,0].set_title("Price Distribution")
axes[0,0].set_xlabel("Price")
axes[0,0].set_ylabel("Frequency")

sns.histplot(df['carat'], kde=True, ax=axes[0,1], bins=30 )
axes[0,1].set_title("Carat Distribution")
axes[0,1].set_xlabel("Carat")
axes[0,1].set_ylabel("Frequency")

sns.histplot(df['depth'], kde=True, ax=axes[1,0], bins=30 )
axes[1,0].set_title("Depth Distribution")
axes[1,0].set_xlabel("Depth (%)")
axes[1,0].set_ylabel("Frequency")

sns.histplot(df['table'], kde=True, ax=axes[1,1], bins=30 )
axes[1,1].set_title("Table Distribution")
axes[1,1].set_xlabel("table (%)")
axes[1,1].set_ylabel("Frequency")

plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```

##### Price Distribution: We can observe that the price distribution is right-skewed, with most diamonds falling in the lower price range.
###### Carat Distribution: The carat weight distribution is also right-skewed, with the majority of diamonds being under 1.5 carats.
##### Depth Distribution: Depth appears to be normally distributed around 60-65%.
##### Table Distribution: The table width distribution seems to center around 55-60%.

```python
# count plot
```

```python
# Setting up the subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Count of Categorical Variables')

# Plotting count plots for categorical variables
sns.countplot(x='cut', data=df, order=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'], ax=axes[0])
axes[0].set_title('Cut Distribution')
axes[0].set_xlabel('Cut')
axes[0].set_ylabel('Count')

sns.countplot(x='color', data=df, order=['J', 'I', 'H', 'G', 'F', 'E', 'D'], ax=axes[1])
axes[1].set_title('Color Distribution')
axes[1].set_xlabel('Color')
axes[1].set_ylabel('Count')

sns.countplot(x='clarity', data=df, order=['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF'], ax=axes[2])
axes[2].set_title('Clarity Distribution')
axes[2].set_xlabel('Clarity')
axes[2].set_ylabel('Count')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

##### Cut Distribution: The majority of the diamonds are categorized as "Ideal," followed by "Premium" and "Very Good." Fewer diamonds fall into the "Good" and "Fair" categories.
##### Color Distribution: The distribution of colors shows that most diamonds are in the middle of the color scale (G, E, F), with fewer diamonds in the extreme categories (D, J).
##### Clarity Distribution: Clarity ranges from "I1" (worst) to "IF" (best). Most diamonds fall into the "SI1," "VS2," and "SI2" categories.

## Bivariate analysis

```python
#scatter plot
```

```python
fig, axes = plt.subplots(1,3, figsize=(15,5))
fig.suptitle("Scatter plot for continous variables vs Price")

sns.scatterplot(x='carat',y='price', data = df, ax=axes[0], alpha=0.5)
axes[0].set_title("Price vs Carat")
axes[0].set_xlabel("Carat")
axes[0].set_ylabel("Price")

sns.scatterplot(x='depth', y='price', data=df, ax=axes[1], alpha=0.5)
axes[1].set_title('Price vs. Depth')
axes[1].set_xlabel('Depth (%)')
axes[1].set_ylabel('Price')

sns.scatterplot(x='table', y='price', data=df, ax=axes[2], alpha=0.5)
axes[2].set_title('Price vs. Table')
axes[2].set_xlabel('Table (%)')
axes[2].set_ylabel('Price')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

##### Price vs. Carat: A clear positive correlation is observed, indicating that as the carat weight increases, the price of the diamond also tends to increase.
##### Price vs. Depth: There doesn't appear to be a strong linear relationship between depth percentage and price.
##### Price vs. Table: Similarly, the table percentage doesn't show a clear linear relationship with price.

```python
# boxplot
```

```python
# Setting up the subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Box Plots of Categorical Variables vs. Price')

# Plotting box plots for categorical variables vs. price
sns.boxplot(x='cut', y='price', data=df, order=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'], ax=axes[0])
axes[0].set_title('Price vs. Cut')
axes[0].set_xlabel('Cut')
axes[0].set_ylabel('Price')

sns.boxplot(x='color', y='price', data=df, order=['J', 'I', 'H', 'G', 'F', 'E', 'D'], ax=axes[1])
axes[1].set_title('Price vs. Color')
axes[1].set_xlabel('Color')
axes[1].set_ylabel('Price')

sns.boxplot(x='clarity', y='price', data=df, order=['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF'], ax=axes[2])
axes[2].set_title('Price vs. Clarity')
axes[2].set_xlabel('Clarity')
axes[2].set_ylabel('Price')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

##### Price vs. Cut: We can observe some variation in price across different cut qualities. Diamonds with "Fair" and "Good" cuts tend to have a higher median price, possibly related to other attributes like carat size.
###### Price vs. Color: There is a noticeable trend where diamonds with colors closer to "J" (lower quality) have a lower median price, while those closer to "D" (higher quality) have a higher median price.
##### Price vs. Clarity: The relationship between clarity and price is not as straightforward. Some higher clarity levels have lower median prices, possibly due to interplay with other factors like carat size and cut quality.

## Outliers

```python
# Setting up the subplots
fig, axes = plt.subplots(1, 4, figsize=(18, 5))
fig.suptitle('Box Plots to Identify Outliers')

# Plotting box plots for potential outlier detection
sns.boxplot(x=df['price'], ax=axes[0])
axes[0].set_title('Price')
axes[0].set_xlabel('')

sns.boxplot(x=df['carat'], ax=axes[1])
axes[1].set_title('Carat')
axes[1].set_xlabel('')

sns.boxplot(x=df['depth'], ax=axes[2])
axes[2].set_title('Depth (%)')
axes[2].set_xlabel('')

sns.boxplot(x=df['table'], ax=axes[3])
axes[3].set_title('Table (%)')
axes[3].set_xlabel('')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
